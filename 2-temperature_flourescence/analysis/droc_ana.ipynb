```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import numpy as np
```

```python
files = glob.glob("../data/*")
files_csv = []

for i in range(len(files)):
    files_csv.append(glob.glob(files[i] + "/*.csv"))
```

```python
fig, ax = plt.subplots()

def ticklabels(y, pos):
    return r'$e^{{{:.0f}}}$'.format(np.log(y))

for i, csv_list in enumerate(files_csv):
    for j, csv in enumerate(csv_list):
        # print(csv)
        df = pd.read_csv(csv)
        x1 = df["response time"]
        y1 = df["response voltage"] + abs(df["response voltage"][0])
        # print(y1)
        x2 = df["signal time"]
        y2 = df["signal voltage"]
        plt.yscale('log', base=np.e)
        plt.semilogy(x1, y1, base=np.e )
        plt.semilogy(x2, y2, base=np.e )
        plt.title("Sports Watch Data")
        plt.xlabel(r'Time ($\mu{}s$)')
        plt.ylabel(r'Voltage (V)')
        # plt.plot(x1, y1 )
        # plt.plot(x2, y2)
        # ax.yaxis.set_major_formatter(mtick.FuncFormatter(ticklabels))
        plt.show()

```
